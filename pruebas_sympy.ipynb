```python
import numpy as np
import sympy
from sympy.integrals.integrals import integrate
```

```python
x = sympy.Symbol('x')
y = sympy.Symbol('y')
z = sympy.Symbol('z')
print(z.__class__ == sympy.core.symbol.Symbol)
np.set_printoptions(precision=2)
#print(dir(ArithmeticError))
```

```python
print(((1/5)*z**2+2*z-1).diff(z))
```

```python
class operador:
    def __init__(self, cord, multi = 1):
        if cord.__class__ == sympy.core.symbol.Symbol:
            self.cord = cord
            self.multi = multi
        else:
            raise TypeError("Coloque una coordenada tipo símbolode sympy, sea serio")
        #fin if 
    #fin funcion
    def __mul__(self, other):
        if "sympy.core" in str(other.__class__):
            return self.multi*other.diff(self.cord)
        elif isinstance(other, (int, float)):
            return 0
        else:
            raise TypeError("Aplique este operador a un objeto tipo sympy.core")
        #fin if 
    #fin función
    def __rmul__(self, other):
        if isinstance(other, (int, float)):
            return operador(self.cord, other)
        else:
            raise TypeError("Por Ahora esta clase solo soporta que se multiplique escalares por detrás")
#fin clase

d_dx = operador(x)
d_dy = operador(y)
d_dz = operador(z)
```

```python
B = [[d_dx, 0, 0], [0, d_dy, 0], [0, 0, d_dz], [0, 0.5*d_dz, 0.5*d_dy], [0.5*d_dz, 0, 0.5*d_dx], [0.5*d_dy, 0.5*d_dx, 0]]
B = np.array(B)
```

```python
phi_minusculo = [1, x, y, z, x*y, x*z, y*z, x**2, y**2, z**2]
phi = np.array([phi_minusculo[:] + [0]*20, [0]*10 + phi_minusculo[:] + [0]*10, [0] * 20 + phi_minusculo[:]])
```

```python
Bphi = B.dot(phi)
C = np.array(list(map(lambda q: list(map(lambda p: sympy.Symbol('C' + str(q) + str(p)), range(6))), range(6))))
```

```python
print(Bphi)
```

```python
print(C)
```

```python
gamma1 = (Bphi.T).dot(C).dot(Bphi)
print(gamma1)
```

```python
phi_min21 = [x, x*y, x*z]
phi_min22 = [1, y, z, y*z, x**2, y**2, z**2]
phi2 = np.array([phi_min21 + [0]*20 + phi_min22, [0]*3 + phi_min22 + [0]*7 + phi_min21 + [0]*10, [0]*10 + phi_min22 + [0]*3 + phi_min21 + [0]*7])
```

```python
Bphi2 = B.dot(phi2)
print(Bphi2)
```

```python
gamma2 = (Bphi2.T).dot(C).dot(Bphi2)
E2 = (phi2.T).dot(phi2)
```
